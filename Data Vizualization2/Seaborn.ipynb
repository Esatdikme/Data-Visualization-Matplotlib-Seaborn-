```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
df=pd.read_csv("diabetes.csv")
df
```

```python
df.info()
```

```python
df.columns
```

```python
df[['Glucose', 'BloodPressure', 'SkinThickness', 'Insulin',
       'BMI', 'DiabetesPedigreeFunction']]=df[['Glucose', 'BloodPressure', 'SkinThickness', 'Insulin',
       'BMI', 'DiabetesPedigreeFunction']].replace(0,np.nan)
df
```

```python
df=df.fillna(value=df.idxmax())
df
```

```python
df.corr()
```

```python
#Heat graph of the data according to the correlation

sns.heatmap(df.corr(),annot=True,cmap="coolwarm",)
```

```python
# Histogram graph of glucose value

sns.histplot(df[["Glucose"]],x="Glucose")
```

```python
# Density graph of glucose value

sns.kdeplot(df.Insulin,shade=True,color="r")
```

```python
# Graph of the binary relations of variables

sns.pairplot(df[['Glucose', 'BloodPressure', 'SkinThickness', 'Insulin',
       'BMI', 'DiabetesPedigreeFunction','Outcome']])
```

```python
# Diabetes graph by the number of pregnancies

sns.countplot(x="Pregnancies",hue="Outcome",data=df[["Pregnancies","Outcome"]])
```

```python
# Diabetes graph of those whose blood pressure is greater than 30

sns.boxplot(x="Outcome",y="BloodPressure", data=df[df["BloodPressure"]>30])
```
